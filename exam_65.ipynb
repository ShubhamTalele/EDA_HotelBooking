## 4.  Hypothesis Testing         

## H₀: There is no difference in ADR between bookings made through Online TA and Direct channels
→ Test: Independent t-test (two groups, numeric target)

```python
from scipy.stats import ttest_ind
group1 = df[df['distribution_channel'] == 'TA/TO']['adr']
group2 = df[df['distribution_channel'] == 'Direct']['adr']
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
print(f"t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
```

```python
p < 0.05	Reject H₀ → ADR differs between channels
p ≥ 0.05	Fail to reject H₀ → ADR not significantly different
`As p-value < 0.05 we Reject H₀ that means there is Significant difference in ADR.
```

## H0: Room upgrades are independent of lead time

```python
from scipy.stats import f_oneway

df['was_upgraded'] = df['assigned_room_type'].astype(str) != df['reserved_room_type'].astype(str)



# Create the two groups
lead_upgraded = df[df['was_upgraded']]['lead_time']
lead_not_upgraded = df[~df['was_upgraded']]['lead_time']

# Run ANOVA
f_stat, p_value = f_oneway(lead_upgraded, lead_not_upgraded)

print(f"F-statistic = {f_stat:.2f}, p-value = {p_value:.4f}")
```

```python
p < 0.05:
→ We reject the null hypothesis.
→ There is a statistically significant difference in lead time between upgraded and non-upgraded bookings.
As p-value 0 < 0.05, we reject the Ho, No evidence that lead time affects upgrades.
```

```python
There is a significant association between room upgrades and lead time. This means the timing of the booking does affect the likelihood of receiving a room upgrade.
```

## H0: Average stay duration does not differ between customer types

```python
from scipy.stats import f_oneway
groups = [group['total_nights'] for name, group in df.groupby('customer_type')]
f_stat, p_value = f_oneway(*groups)
print(f"F-statistic = {f_stat:.2f}, p-value = {p_value:.4f}")
```

```python
There is a significant difference in average stay duration across different customer types.
This means customer type influences how long people stay
As p-value 0 < 0.05, we reject the Ho, no significant difference in stay duration
```
